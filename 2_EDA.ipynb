```python
# Crea una grfica que muestre la cantidad de sobrevivientes por puerto de embarque (Embarked).

plt.figure(figsize = (12, 8))

# Paleta de colores.
color = ['#240750', '#344C64']

# Creardo el countplot y agregando los datos en la parte superior de las barras.
ax = sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)

# Personalizacin.
plt.legend(title = 'Supervivencia', labels = ['No sobrevivi', 'S sobrevivi'])
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
# Agrupa las edades en categoras (Infante, Nio, Adolescente, Joven Adulto, Adulto, Mayor) y crea una grfica que muestre la tasa de 
# supervivencia por grupo de edad.

# Definir categoras de edad
bins = [0, 2, 12, 18, 30, 50, 100]  # Rangos de edad.
labels = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']

# Crear columna con grupos de edad.
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Calcular la tasa de supervivencia por grupo.
survival_rate = df.groupby('AgeGroup', observed=True)['Survived'].mean().reset_index()

# Multiplicar por 100 para porcentaje.
survival_rate['Survived'] = survival_rate['Survived'] * 100

# Crear la grfica.
plt.figure(figsize=(10, 6))
sns.barplot(
    data=survival_rate,
    x='AgeGroup',
    y='Survived',
    hue='AgeGroup',
    palette="viridis",
    legend=False
)

# Ttulos y ejes.
plt.title("Tasa de Supervivencia por Grupo de Edad (%)", fontsize=16)
plt.xlabel("Grupo de Edad")
plt.ylabel("Tasa de Supervivencia (%)")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Etiquetas en cada barra.
for i, row in survival_rate.iterrows():
    plt.text(i, row['Survived'] + 1, f"{row['Survived']:.1f}%", ha='center', fontsize=10)

plt.show()
```

```python
# Crea una grfica que muestre el conteo de pasajeros en cada clase (Pclass).

plt.figure(figsize=(12, 7))

# Creardo el countplot y agregando los datos en la parte superior de las barras.
ax = sns.countplot(data=df, x='Pclass')

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, padding=3)

# Personalizacin.
plt.title("Conteno de pasajeros por Clase", fontsize=16)
plt.xlabel("Clase", fontsize=12)
plt.ylabel("Total", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
# Encuentra y muestra los valores mximo y mnimo de la columna de edad.

max_age = df['Age'].max()
min_age = df['Age'].min()

print("Edad minima:", min_age)
print("Edad maxima:", max_age)
```

```python
# Crea una grfica que muestre la distribucin de edades, incluyendo lneas verticales que indiquen la media, mediana y moda de las edades.

plt.figure(figsize = (12, 8))

# Crear hisplot de 'Age'.
sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribucin de Edades')

# Personalizacion y agregando las lineas.
line_media = plt.axvline(df['Age'].mean(), color='red', label=f"Media: {df['Age'].mean():.2f}")
line_mediana = plt.axvline(df['Age'].median(), color='black', label=f"Mediana: {df['Age'].median():.2f}")
line_moda = plt.axvline(df['Age'].mode()[0], color='green', label=f"Moda: {df['Age'].mode()[0]:.2f}")

plt.legend(handles=[line_media, line_mediana, line_moda], title="Medidas de Tendencia Central")

plt.show()
```

```python
# Crea un diagrama de caja (boxplot) que muestre los outliers en la columna de edad.

plt.figure(figsize=(8, 6))

# Crear boxplot de la columna 'Age'
sns.boxplot(data=df, y='Age', color='skyblue')

# Personalizacin del grfico.
plt.title("Diagrama de Caja de Edad con Outliers", fontsize=16)
plt.ylabel("Edad")
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
